# Imports

```python
import matplotlib.pyplot as plt 
import numpy as np 
from scipy.interpolate import CubicSpline, interp1d 
```

```python
end = 10
start = -10
steps = 1
```

```python
plt.rcParams['figure.figsize'] =(12,8) 

xs = np.arange(start, end, 0.01) 
x = np.arange(start,end+1, steps) 
y = 1/(1+25*(x**2)) 
```

# Apply linear interpolation 

```python
linear_int = interp1d(x, y) 
ys_lin = linear_int(xs) 
```

# Apply cubic spline interpolation 


```python
cs = CubicSpline(x, y) 
```

# Apply polynomial interpolation 


```python
coeffs = np.polyfit(x, y, deg=len(x)-1)
ys_poly = np.polyval(coeffs, xs)
```

# Plot linear interpolation 

```python
plt.plot(x, y, 'o', label='data') 
plt.plot(xs, ys_lin, label="Linear", color='green') 
plt.plot(xs, 1/(1+25*(xs**2)), label='true') 
plt.legend(loc='upper right', ncol=2) 
plt.title('Linear Interpolation') 
plt.show() 
```

# Plot cubic spline interpolation 

```python
plt.plot(x, y, 'o', label='data') 
plt.plot(xs, 1/(1+25*(xs**2)), label='true') 
plt.plot(xs, cs(xs), label="Cubic Spline") 
plt.ylim(-1.5, 1.5) 
plt.legend(loc='upper right', ncol=2) 
plt.title('Cubic Spline Interpolation') 
plt.show()
```

# Plot polynomial interpolation

```python
plt.plot(x, y, 'o', label='data')
plt.plot(xs, 1/(1+25*(xs**2)), label='true')
plt.plot(xs, ys_poly, label="Polynomial")
plt.legend(loc='upper right', ncol=2)
plt.title('Polynomial Interpolation')
plt.show()
```
